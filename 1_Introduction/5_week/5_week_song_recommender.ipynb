# Song Recommender

```python
import graphlab as gl
gl.canvas.set_target('ipynb')
```

```python
song_data = gl.SFrame('Data/song_data.gl/')
```

### 1. Explore Data

```python
print len(song_data)
song_data.head()
```

```python
song_data['title'].show()
```

### 2. Look at Users

```python
## look at unique users 
users = song_data['user_id'].unique()
print len(users)
```

### 3. Create a song recommender 

```python
## split training and test data
train_data,test_data = song_data.random_split(.8,seed=0)
```

##### 3.1 Popularity model -- recommend only by popularity

```python
## simple popularity-based recommender
popularity_model = gl.popularity_recommender.create(train_data,user_id='user_id',item_id='song')
```

```python
## use popularity model to make predictions
popularity_model.recommend(users=[users[0]])      # this argument take a list 
```

```python
## use popularity model to make predictions --- using popularity model, everybody get the same thing
popularity_model.recommend(users=[users[1]])
```

#### 3.2 Build a song recommender with personalization 

```python
## using item similarity model 
personalized_model = gl.item_similarity_recommender.create(train_data,user_id='user_id',item_id='song')
```

```python
## use this model to make recommendations
personalized_model.recommend(users=[users[0]])
```

```python
## use this model to make recommendations for user 2 
personalized_model.recommend(users=[users[1]])
```

```python
## using this, we can also get similar items 
personalized_model.get_similar_items(['With Or Without You - U2'])
```

## 4. Evaluate and compare two models 

```python
%matplotlib inline
```

```python
model_performance = gl.recommender.util.compare_models(test_data,
                                                       [popularity_model,personalized_model],
                                                      user_sample=0.05) ## put sample to be 0.05, becauser of computer constrains
```
